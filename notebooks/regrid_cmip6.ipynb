```python
import matplotlib.pyplot as plt 
model_names = list(processed_datasets.keys())
n_models = len(model_names)

# Crear una cuadrícula de subplots
fig, axes = plt.subplots(nrows=n_models, ncols=1, figsize=(5 ,5 * n_models), squeeze=False)

# Iterar sobre los modelos y graficar en cada subplot
for i, model_name in enumerate(model_names):
    ax = axes[i, 0]  # Acceder al subplot correspondiente
    regrid_data[model_name].sel(param='D1').shdr.isel(time=200).plot(ax=ax)
    ax.set_title(model_name)  # Establecer el título con el nombre del modelo

# Ajustar los espacios entre subplots
plt.tight_layout()
plt.show()
```
